Lasso

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Diccionari amb fitxers i colors
models = {
    'TF-idf': ('data/lasso_TF-idf_log.pkl', 'dodgerblue'),
    'W2V':    ('data/lasso_W2V_log.pkl',    'forestgreen'),
    'TB':     ('data/lasso_TB_log.pkl',     'darkorange'),
    'NV':     ('data/lasso_NV_log.pkl',     'firebrick')
}

val_scores_by_model = {}
val_mse_by_model = {}

# Gràfic 1: R² Score
plt.figure(figsize=(12, 7))
for name, (path, color) in models.items():
    df = pd.read_pickle(path)
    x_axis = df.iloc[:,0]
    train = df.iloc[:,1]
    val = df.iloc[:,2]

    val_scores_by_model[name] = val

    plt.plot(x_axis, train, label=f"{name} - Train R²", color=color, linestyle='-')
    plt.plot(x_axis, val, label=f"{name} - Val R²", color=color, linestyle='--', alpha=0.7)

plt.xscale('log')
plt.xlabel('alpha')
plt.ylabel('R² Score')
plt.title('Lasso: Train vs Validation R² per representació')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Gràfic 2: MSE
plt.figure(figsize=(12, 7))
for name, (path, color) in models.items():
    df = pd.read_pickle(path)
    x_axis = df.iloc[:,0]
    val_mse = df.iloc[:,3]

    val_mse_by_model[name] = val_mse

    plt.plot(x_axis, val_mse, label=f"{name} - Val MSE", color=color, linestyle='-', marker='o')

plt.xscale('log')
plt.xlabel('alpha')
plt.ylabel('Mean Squared Error')
plt.title('Lasso: Validation MSE per representació')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Resultats

```python
import pandas as pd

# Models i transformacions
models = ['reg','lasso', 'ridge', 'rf', 'xgb', 'svr']
transformacions = ['TF-idf', 'W2V', 'TB', 'NV']

# Inicialitzar els diccionaris per emmagatzemar les mètriques
train_scores = []
val_scores = []
val_mse = []

# Crear les llistes buides per a les taules
train_scores_data = []
val_scores_data = []
val_mse_data = []

# Llegir els fitxers per cada model i transformació i obtenir les mètriques
for model in models:
    train_scores_row = []
    val_scores_row = []
    val_mse_row = []
    
    for transformacio in transformacions:
        # Generar el path per a cada fitxer
        path = f"data/{model}_{transformacio}_log_rem_opt.pkl"
        
        # Llegir el fitxer pickle
        df = pd.read_pickle(path)
        
        # Obtenir les mètriques
        train_score_mean = df['train_score_mean'].iloc[0]
        val_score_mean = df['val_score_mean'].iloc[0]
        val_mse_mean = df['val_mse_mean'].iloc[0]
        
        # Afegir les mètriques a les files
        train_scores_row.append(train_score_mean)
        val_scores_row.append(val_score_mean)
        val_mse_row.append(val_mse_mean)
    
    # Afegir les files a les taules
    train_scores_data.append(train_scores_row)
    val_scores_data.append(val_scores_row)
    val_mse_data.append(val_mse_row)

# Crear les taules amb models com a files i transformacions com a columnes
df_train_scores = pd.DataFrame(train_scores_data, columns=transformacions, index=models)
df_val_scores = pd.DataFrame(val_scores_data, columns=transformacions, index=models)
df_val_mse = pd.DataFrame(val_mse_data, columns=transformacions, index=models)

# Mostrar les taules
print("Train Scores:")
print(df_train_scores)
print("\nValidation Scores:")
print(df_val_scores)
print("\nValidation MSE:")
print(df_val_mse)

# Opcional: guardar les taules en fitxers CSV o pickle si ho desitges
# df_train_scores.to_csv('train_scores.csv', index=True)
# df_val_scores.to_csv('val_scores.csv', index=True)
# df_val_mse.to_csv('val_mse.csv', index=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Models i transformacions
models = ['reg','lasso', 'ridge', 'rf', 'xgb', 'svr']
transformacions = ['TF-idf', 'W2V', 'TB', 'NV']

# Colors que has definit per cada transformació
colors = {
    'TF-idf': 'dodgerblue',
    'W2V': 'forestgreen',
    'TB': 'darkorange',
    'NV': 'firebrick'
}

# Inicialitzar la figura per al gràfic
plt.figure(figsize=(12, 7))

# Recórrer els models i les transformacions per a crear els punts
for model in models:
    for transformacio in transformacions:
        # Generar el path per a cada fitxer
        path = f"data/{model}_{transformacio}_log_rem_opt.pkl"
        
        # Llegir el fitxer pickle
        df = pd.read_pickle(path)
        
        # Obtenir els valors de training i testing scores
        train_score = df['train_score_mean'].iloc[0]
        val_score = df['val_score_mean'].iloc[0]
        
        # Crear el gràfic de punts
        # Punts per al training score
        plt.scatter(transformacio, train_score, color=colors[transformacio], label=f"{model} - Train {transformacio}" if model == 'lasso' else "", marker='o', s=100, edgecolors='black', alpha=0.7)
        
        # Punts per al testing score
        plt.scatter(transformacio, val_score, color=colors[transformacio], label=f"{model} - Val {transformacio}" if model == 'lasso' else "", marker='x', s=100, edgecolors='black', alpha=0.7)

# Afegir les llegendes i etiquetes
plt.xlabel('Transformació')
plt.ylabel('R² Score')
plt.title('Training vs Testing R² Score per Transformació i Model')
plt.xticks(transformacions)  # Per mostrar les transformacions a l'eix X
plt.legend(title="Models", bbox_to_anchor=(1.05, 1), loc='upper left')

# Afegir una graella
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gràfic
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Models i transformacions
models = ['reg','lasso', 'ridge', 'rf', 'xgb', 'svr']
transformacions = ['TF-idf', 'W2V', 'TB', 'NV']

# Colors que has definit per cada transformació
colors = {
    'TF-idf': 'dodgerblue',
    'W2V': 'forestgreen',
    'TB': 'darkorange',
    'NV': 'firebrick'
}

# Inicialitzar la figura per al gràfic
plt.figure(figsize=(12, 7))

# Recórrer els models i les transformacions per a crear els punts
for model in models:
    for transformacio in transformacions:
        # Generar el path per a cada fitxer
        path = f"data/{model}_{transformacio}_log_rem_opt.pkl"
        
        # Llegir el fitxer pickle
        df = pd.read_pickle(path)
        
        # Obtenir els valors de validation MSE
        val_mse = df['val_mse_mean'].iloc[0]
        
        # Crear el gràfic de punts per MSE (usant x per testing MSE)
        plt.scatter(transformacio, val_mse, color=colors[transformacio], 
                    label=f"{model} - Val MSE {transformacio}" if model == 'lasso' else "", 
                    marker='x', s=100, edgecolors='black', alpha=0.7)

# Afegir les llegendes i etiquetes
plt.xlabel('Transformació')
plt.ylabel('Mean Squared Error')
plt.title('Validation MSE per Transformació i Model')
plt.xticks(transformacions)  # Per mostrar les transformacions a l'eix X
plt.legend(title="Models", bbox_to_anchor=(1.05, 1), loc='upper left')

# Afegir una graella
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gràfic
plt.tight_layout()
plt.show()
```

Diferències significatives del R2 validation

```python
from scipy.stats import f_oneway
import pandas as pd

# Suposant que tens les taules amb els R² de validació per a cada transformació i model.
# Afegir les dades del R² per a cada model per cada transformació a una llista
r2_values = {
    'TF-idf': [],
    'W2V': [],
    'TB': [],
    'NV': []
}

# A partir de les taules df_val_scores, extraiem les mètriques
# Aquí suposem que tens el df_val_scores on les columnes són les transformacions i les files els models

# Afegir les dades de cada transformació
for transformacio in ['TF-idf', 'W2V', 'TB', 'NV']:
    r2_values[transformacio] = df_val_scores[transformacio].values

# Realitzar el test ANOVA de una via
f_statistic, p_value = f_oneway(r2_values['TF-idf'], r2_values['W2V'], r2_values['TB'], r2_values['NV'])

# Imprimir els resultats
print("Estadístic F:", f_statistic)
print("Valor p:", p_value)

# Comprovar si el valor p és significatiu
if p_value < 0.05:
    print("Hi ha diferències significatives entre les transformacions.")
else:
    print("No hi ha diferències significatives entre les transformacions.")
```
